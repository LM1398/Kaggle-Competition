```python
# Q3: 所得は、以下のように決定されるとする。
#     0 ~ 15歳未満 >>> 0万円
#     15 ~ 23歳未満 >>> 0万円 ~ 100万円の間の整数。年齢によらずランダムに生成される。
#     23 ~ 50歳未満 >>> 200万円 × 年齢 + ε1
#                      ε1は[-20, -10, 0, 10, 20]のいずれかの値をランダムに取る確率変数。
#                      HINT: np.random.choiceを使用せよ。
#     51 ~         >>> -10万円 × 年齢 +1000 + ε2
#                      ε2は[0, 10, 20]のいずれかの値をランダムに取る確率変数
#                      HINT: np.random.choiceを使用せよ。
#
#     また、性別が以下のように所得に影響を与えるものとする。
#     LGBTならば、-100万円
#     女性ならば、-50万円
#     男性ならば、100万円
#     （※ 所得に関しては、依然として男性が高い。）
#     0はLGBT、1は女性、2は男性とする。
#
#     なお、性別の効果で所得が負の値を取る結果となってしまうときは、所得は0になるように調整する。
#     HINT: np.maxを使用せよ。
#
#     これらの条件から1000人分の所得のndarrayを生成せよ。
```

```python
income = []

dice = [-20, -10, 0, 10, 20]
dice2 = [0, 10, 20]

for x in age:
    if x < 15:
        income.append(0)
    elif 15 <= x < 23:
        income.append(np.random.choice(range(0,101)))
    elif 23 <= x < 50:
        income.append(x * 200 + np.random.choice(dice))
    else:
        income.append(x * -10 + 1000 + np.random.choice(dice2))  
```

```python
income_sex = []
for x in sex:
    if x == 0:
        income_sex.append(-100)
    elif x == 1:
        income_sex.append(-50)
    else: income_sex.append(100)
```
